```python
def visualizer(model,image):
  result = inference_detector(model, image)
  detections = sv.Detections.from_mmdetection(result)
  detections = detections[detections.confidence > 0.25].with_nms()
  box_annotator = sv.BoxAnnotator()
  labels = [
      f"{classes[class_id]} {confidence:0.2f}"
      for _, _, confidence, class_id, _
      in detections
  ]
  annotated_image = box_annotator.annotate(image.copy(), detections, labels=labels)
  return annotated_image
```

```python
def callback(image: np.ndarray) -> sv.Detections:
    result = inference_detector(model, image)
    detections = sv.Detections.from_mmdetection(result)
    detections.class_id = np.array(sorted(detections.class_id))
    return detections[detections.confidence > 0.2].with_nms(threshold=0.5)


confusion_matrix = sv.ConfusionMatrix.benchmark(
    dataset = ds,
    callback = callback
)

_ = confusion_matrix.plot()
```

```python
mean_average_precision = sv.MeanAveragePrecision.benchmark(
    dataset = ds,
    callback = callback
)

print('mAP:', mean_average_precision.map50_95)
```
