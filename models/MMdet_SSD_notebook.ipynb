```python
import time
import numpy as np
from mmdet.apis import inference_detector

# Ensure the model is loaded and the device is set correctly
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")

# Path to the image
image_path = "/content/drive/MyDrive/task_03/turning_dataset/val/images/00538.jpg"
image = cv2.imread(image_path)

# Warm-up iterations to ensure everything is loaded properly
warmup_iterations = 10
for _ in range(warmup_iterations):
    _ = inference_detector(model, image)

# Number of iterations to run for averaging FPS
num_iterations = 100

# List to store time taken for each iteration
times = []

# Run inference multiple times to calculate average time
for _ in range(num_iterations):
    start_time = time.time()
    _ = inference_detector(model, image)
    end_time = time.time()
    times.append(end_time - start_time)

# Calculate average time taken for inference
average_time = np.mean(times)

# Calculate FPS
fps = 1 / average_time

print(f"Average FPS: {fps:.2f}")
```

```python
def sort(confidence):
  #confidence  = detections.class_id
  new_conf = []
  for i in confidence:
    if i == 2:
      new_conf.append(int(0))
    elif i == 0:
      new_conf.append(int(2))
    else:
      new_conf.append(i)
  return np.array(new_conf)
```

```python
def callback(image: np.ndarray) -> sv.Detections:
    result = inference_detector(model, image)
    detections = sv.Detections.from_mmdetection(result)
    detections.class_id = sort(detections.class_id)
    return detections[detections.confidence > 0.2].with_nms(threshold=0.5)


confusion_matrix = sv.ConfusionMatrix.benchmark(
    dataset = ds,
    callback = callback
)

_ = confusion_matrix.plot()
```

```python
mean_average_precision = sv.MeanAveragePrecision.benchmark(
    dataset = ds,
    callback = callback
)

print('mAP:', mean_average_precision.map50_95)
```

```python
per_class_map = mean_average_precision.per_class_ap50_95.mean(axis=1)
for class_name, value in zip(ds.classes, per_class_map):
    print(f"{class_name}: {value:.2f}")
```
